# Summary of the Breast Cancer Wisconsin Dataset:

The Breast Cancer Wisconsin dataset is a collection of clinical data used to predict the presence of breast cancer. It consists of 569 instances of breast cancer cases, each described by 30 numerical features computed from a digitized image of a fine needle aspirate (FNA) of a breast mass. The features include measurements such as radius, texture, perimeter, area, and smoothness, each computed for mean, standard error, and worst or largest value. The outcome variable, diagnosis, is a binary classification indicating whether the tumor is malignant (M) or benign (B).

# Explatory data analysis (EDA)


Let's perform Exploratory Data Analysis (EDA) on the breast cancer dataset to gain insights for further analysis or modeling. 

```python
# Check the distribution of the diagnosis variable
diagnosis_counts = df['diagnosis'].value_counts()
print(diagnosis_counts)
```

Based on the data above, it looks like we have 357 benign tumors and 212 malignant ones.

```python
#Lets's first perform a styled summary of the dataset's descriptive statistics using Pandas and its styling capabilities.
#To visually represent the mean values for each feature
#Visualizing the standard deviation values, where higher values have a more intense red color.
#show the distribution of median values with a gradient from cool to warm colors.
df.describe().T.style.bar(subset=['mean'], color='lightblue')\
                        .background_gradient(subset=['std'], cmap='Purples')\
                        .background_gradient(subset=['50%'], cmap='PiYG')
```

```python
# Bar plot for diagnosis distribution
sns.countplot(x='diagnosis', data=df)
plt.title('Diagnosis Distribution')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['diagnosis'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'salmon'], labels=['Benign', 'Malignant'])
plt.title('Diagnosis Distribution')
plt.ylabel('')
plt.show()
```

```python
# Histograms of features
df.hist(bins=20, figsize=(20, 15))
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(20, 18))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```

```python
# List of 'mean' features and 'diagnosis'
mean_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 
                 'smoothness_mean', 'compactness_mean', 'concavity_mean', 
                 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']
# Create a DataFrame with the selected features
df_mean = df[mean_features]
# Descriptive statistics for 'mean' features
se_stats = df_mean.describe()
print(se_stats)
```

```python
# Box plots for "mean" features by diagnosis
for feature in mean_features:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='diagnosis', y=feature, data=df, palette="muted")
    plt.title(f'{feature} by Diagnosis')
    plt.show()
```

```python
# List of 'se' features and 'diagnosis'
se_features = ['radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 
               'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 
               'fractal_dimension_se', 'diagnosis']

# Create a DataFrame with the selected features
df_se = df[se_features]
# Descriptive statistics for 'se' features
se_stats = df_se.describe()
print(se_stats)
```

```python
# Box plots for "se" features by diagnosis
for feature in se_features[:-1]:  # Exclude 'diagnosis' from the feature list
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='diagnosis', y=feature, data=df_se)
    plt.title(f'{feature} by Diagnosis')
    plt.show()
```

```python
# List of 'worst' features and 'diagnosis'
worst_features = ['radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 
                  'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 
                  'fractal_dimension_worst', 'diagnosis']

# Create a DataFrame with the selected features
df_worst = df[worst_features]
# Descriptive statistics for "worst" features
worst_stats = df_worst.describe()
print(worst_stats)
```

```python
# Box plots for "worst" features by diagnosis
for feature in worst_features[:-1]:  # Exclude 'diagnosis' from the feature list
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='diagnosis', y=feature, data=df_worst)
    plt.title(f'{feature} by Diagnosis')
    plt.show()
```
